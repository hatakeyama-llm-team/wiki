```python
path="dump/jawikibooks-20240320-pages-articles-multistream.xml"
```

```python
import xml.etree.ElementTree as ET

# XMLファイルを読み込む
tree = ET.parse(path)
root = tree.getroot()
```

```python
print(root.tag)
```

```python
for child in root:
    print(child.tag, child.attrib)
```

```python
import glob
import unicodedata
import re
from bs4 import BeautifulSoup

tag="{http://www.mediawiki.org/xml/export-0.10/}page"
elems=root.findall(tag)
n_data=200
start_id=300
cnt=0
for elem in elems:
    cnt+=1
    if cnt<start_id:
        continue
    if cnt>n_data+start_id:
        break

    for elem2 in elem:
        #print(elem2.text)
        for elem3 in elem2:
            text=elem3.text
            if text is None:
                continue

            text = re.sub(r'[“”]', '', text)
            text = re.sub(r'https?:\/\/.*?[\r\n ]', '', text)
            soup = BeautifulSoup(text, 'html.parser')
            text = soup.get_text()
            text=text.replace("\n\n","\n")

            # 正規表現で[[w:任意のテキスト|任意のテキスト]]形式を探して置換
            text = re.sub(r'\[\[w:.*?\|(.*?)\]\]', r'\1', text)
            # 正規表現で{{任意のテキスト}}形式を探して削除
            text= re.sub(r'\{\{.*?\}\}', '', text)
            text= re.sub(r'\{.*?\}', '', text)
            # 正規表現で'''任意のテキスト'''形式を探して置換
            text= re.sub(r"'''(.*?)'''", r'\1', text)

            # 正規表現で[[...: ....]]形式を探して削除
            text= re.sub(r'\[\[.*?:.*?\]\]', '', text)

            text= re.sub(r'\[\/.*?\]', '', text)


            # 正規表現で[[任意のテキスト]]形式を探して置換
            text= re.sub(r'\[\[(.*?)\]\]', r'\1', text)
            
            # 正規表現で日付、日時で始まる行を探して削除
            text= re.sub(r'^\d{4}年\d{1,2}月\d{1,2}日 \(.\) \d{2}:\d{2}.*\n?', '', text, flags=re.MULTILINE)
            if len(text)<100:
                continue
            
            text=unicodedata.normalize('NFKC', text)

            print(text)
            print("---\n\n")
```
